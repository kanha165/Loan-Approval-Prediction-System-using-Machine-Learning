```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
```

```python
model = joblib.load("loan_approval_model.pkl")
print("Model loaded successfully")
```

```python
test_df = pd.read_csv("test.csv")
test_df.head()
```

```python
test_df.drop("Loan_ID", axis=1, inplace=True)
```

```python
test_df.isnull().sum()
```

```python
test_df["LoanAmount"].fillna(test_df["LoanAmount"].mean(), inplace=True)
test_df["Credit_History"].fillna(test_df["Credit_History"].mode()[0], inplace=True)
test_df.fillna(test_df.mode().iloc[0], inplace=True)
```

```python
le = LabelEncoder()

for col in test_df.select_dtypes(include="object"):
    test_df[col] = le.fit_transform(test_df[col])
```

```python
test_df = test_df[model.feature_names_in_]
```

```python
test_predictions = model.predict(test_df)
print("Prediction completed")
```

```python
result = pd.DataFrame({
    "Loan_Status_Predicted": test_predictions
})
```

```python
result["Loan_Status_Predicted"] = result["Loan_Status_Predicted"].map({
    1: "Approved",
    0: "Rejected"
})
```

```python
result.to_csv("loan_test_results.csv", index=False)
print("loan_test_results.csv saved successfully")
```
